# <a id="LearningProb"></a> The learning problem
The are many definitions of the learning problem, each from a slightly different perspective and all very similar. I present below the definition that I selected to use. Tthis definition is somewhat biased towards the statistical viepoint where we consider distributions of the data points. All the distributions considered are assumed to be **unknown**  and only serve as a vehicle for deriving some parts of the learning theory. The elements of the learning problem are:

* ### <font color="green">The data generator </font>
Generates $N$ input points $\mathbf{x_i} \in \mathcal{X} \subset \mathbb{R}^p\quad, i =1,\ldots,N$, supposed to independently and idendically distributed according to an **unknown** fixed distribution $P_X(x)$. The $\mathbf{x_i}$'s are also called **instances** (and then $\mathcal{X}$ is called the instance space). The coordinates of $\mathbf{x_i}$ are called
**features**, **predictors** or **attributes**.

* ### <font color="green">The target operator</font>
Transforms input values into a **response** $y \in \mathcal{Y}$ according to an **unknown** conditional probability distribution $P_{Y\mid X}(y\mid x)$. The learning probelm is called a **regression** problem when $\mathcal{Y} \subset \mathbb{R}$ and a **classification** problem. In the layter case, $\mathcal{Y}$ will sometiomes be called the **label** set and $y$ will be called the label.

    A special case is when the target operator is realized by some deterministic function $y=f(x)$ or a stochastic function $y=f(x)+\epsilon$ where $\epsilon$ is a noise term with zero mean. 
    
    When the function $f(x)$ is boolean we will sometime call it a **concept** and refer to the learning problem as **concept learning**

* ### <font color="green">The training set</font>
$\mathcal{D} = \{(\mathbf{x_i},y_i) ,i = 1,\ldots,N\}$ where  $(\mathbf{x_i},y_i)$ are independetly and identically  distributed according to the **unknown**  joint distribution $P_{X,Y}(x,y)=P_{Y\mid X}(y\mid x)P_X(x)$. Elements of training set are referred to as **samples** or **examples**.

* ### <font color="green">A learning machine</font>
An algorithm which, on the basis of the training set, returns a **prediction rule** $h : \mathcal{X} \rightarrow \mathcal{Y}$ that estimates the value of the target function for each value of an input point $\mathbf{x}$. This function is called a **predictor**, **hypothesis** or a **classifier**. Learning machine is also often referred to as just a **learner**, 

Intuitivly we would like the hypothesis $h$ to be "close" to the function $f$ or. More generally we would like $(\mathbf{x},h(\mathbf{x}))$ to the close joint distribution $P_{X,Y}(x,y)=P_{Y\mid X}(y\mid x)P_X(x)$ on **new**, unseen, data from $\mathcal{X}$. 


# <span id="AnatomyOfLearningMachine">Anatomy</span> of a learning machine

In the paper **"A Few Useful Things to Know about Machine Learning"** ([download](http://homes.cs.washington.edu/~pedrod/papers/cacm12.pdf)) Pedro Domingos argue that any learning machine can be considered as being composed of just three components: **Representation**, realized in the form of an **Hypothesis set (or space)**, **Evaluation** manifested by the **loss function** and **optimization** procedure for finding the best candidate out of the hypothesis set that minimizes the loss function.

* ### <font color="green">An hypothesis set</font>
A set $\mathcal{H}$ of predictors $h : \mathcal{X} \mapsto \mathcal{Y}$ from which we select a particular estimator to the target operator. Many times the selection will be performed by minimizing the empirical risk, a term described below.

* ###<font color="green">The evaluation function and the loss function</font>
An evaluation function, also called **objective** or **scoring** function is a function that is used to rank different hypetheses from the hipothesis set. Many evaluation functions are proposed in the literature including [Precision and recall](http://en.wikipedia.org/wiki/Precision_and_recall), [likelihood](http://en.wikipedia.org/wiki/Maximum_likelihood). [posteriori probability](http://en.wikipedia.org/wiki/Maximum_a_posteriori_estimation), [information gain](http://www.autonlab.org/tutorials/infogain.html), [K-L divergence](http://en.wikipedia.org/wiki/Kullback%E2%80%93Leibler_divergence) and [margin](http://en.wikipedia.org/wiki/Margin_classifier) to name a few.

    I will concentrate below on a special type of evaluation functions called **loss function**. A loss function is a non-negative function $L(h({\mathbf x}),y))$ defined for a function $h : \mathcal{X} \mapsto \mathcal{Y}$ that measure the discrepancy between the value predicted by $h$ for point ${\mathbf x}$ and $y$. Loss function is sometimes also called a **utility** , **penalty** or **cost** function.

    The loss function is used to define two quantities

    * **The expected risk** defined as
       > $R(g) = \int L(g(\mathbf x),y) P_{X,Y}(x,y)dxdy$ 

      We will sometimes refer to $R(h)$ as the **out-of-sample**, **test**, **generalization** or **true** error for a predictor $h$ and denote it by $E_{\tt{out}}$.
      
      The ideal estimator $g_o$ is defined by
      > $g_o = \tt{arg min}_{g \in \mathcal{G}} R(g)$

      where $\mathcal{G}$ is the space of measurable functions for which $R(g)$ is well defined. In some cases the loss function [have specific semantics](http://hunch.net/?p=269) which allow us to precisely characterize $g_o$. For example, for a square loss function $L(h({\mathbf x}),y))=(h(x)-y)^2$ we have that $g_o(x)=\mathbb{E}( Y\mid X= x)$, Thus the best prediction of $Y$ for any $X=x$ is the conditional mean. 
      
      Note that in the special case where the learning problem is formulated in terms of a target function $f$ then $g_o = f$

    * **The empirical risk**  is the risk computed on the data set
      > $R_{\tt emp}(h) = \frac{1}{N}\sum_{i=1}^N L(h(\mathbf x_i),y_i)$  

      The empirical risk is sometiomes also referred to as the **in-sample** or **training**  error and denoted by $E_{\tt in}$ 

 The loss function determine the nature of the learning algorithm and is thus of great interest. In <a href="#LossFunctionSelection" data-ajax="false">this section below</a> we further discuss the question selecting the loss function and present some popular alternatives.

* ### <font color="green">The learning algorithm - Empirical risk minimization</font>
Given an hypothesis set and a loss function, it is natural to search for the "best" candidate from the hypothesis set, that is $g_o\in \mathcal{H}$ so that 
>$g_o = \tt{arg min}_{h \in \mathcal{H}} R_{\tt emp}(h)$

    A central problem is to find conditions under which $g$ mimics the behavior of $g_o$. The approximation of $g_o$ from a finite set of data is an ill posed problem which we address by using [regularization theory](http://en.wikipedia.org/wiki/Regularization_(mathematics). This is usually manifasted by adding smoothness constraints on the solution, that is we instead of minimizing the empirical risk we minimize
    > $\frac{1}{N}\sum_{i=1}^N L(h(\mathbf x_i),y_i) + \lambda J(h)$ 

    where $J(h)$ is a smoothness functional. Typical example is when the hypothesis set $\mathcal{H}$ is he set of linear functions of the form $h=\mathbf{ w^T x}$. In this case, $J(h) = ||\mathbf{w}||^2$
    
* ### <font color="green">The optimization method</font>
  Once a loss function is specified we need to search the hypothesis space for the hypothesis that minimizes the loss. Optimization for machine learning is a very active research area and there are many methods investigated depending on the type of learning problem and nature of the loss function. There is a [Workshop](http://opt-ml.org) dedicated to optimization for machine learning and great part of most machine learning algorithms are devoted to efficient implementations of optimization techniques. 

# <a id="overfitting"></a>Overfitting

The whole purpose of machine learning is to find a learner that performs well on yet **unseen**. [Overfitting](http://en.wikipedia.org/wiki/Overfitting) is the most feared obstacle to learning. It is a comperative terms comparing the performance of two competing models. The problem is that looking at the in-sample errors we think that we are doing extremely well while the out-of-sample error actually increases. In the extreme case, we might have a model that provide a predictor that attains a zero in-sample error but its performance on the test set is much worse than another model which attains some error on the training set but much lower error on th etest set (we will se a computational example below).

In the case of overfitting it is 
[Inductive bias](http://en.wikipedia.org/wiki/Inductive_bias) that comes to our rescue. The idea is that we use our knowledge of the domain to properly select the hypothesis set and the loss function. This shoulde be done **without** viewing the data (bacause if we view the data we will tune our selection to the samples that we see and thus we will be likely to overfit). An extreme case of inductive bias is when we know exactly what program to write to predict. Thus the hypothesis set will consist of just one fixed function. As [Eliezer Yudkowsky note](http://lesswrong.com/lw/hg/inductive_bias/): *"Inductive biases can be probabilistically correct or probabilistically incorrect, and if they are correct, it is good to have as much of them as possible, and if they are incorrect, you are left worse off than if you had no inductive bias at all."*. 

The purpose of the theory of learning is to provide us tools to help in selecting the hypothesis set and loss function and to give us some idea as to what size of training set do we need to get descent generalization error. As we will see, learning theory actually guarantees that no single method consistently outperforms other methods. Thus we need deep domain knowledge and rich set of learning methods so we can hope to get a method that is good for our problem. We need to accept that it well may be that no existing method provide good generalization for our problem and we will have to work hard on developing such a method.


## <a id="overfittingExamplePolynomial"></a>Example
We will consider 12 points sampled from our test function which is a polynomial of degree 15, plotted from now on io a <font color="blue">**blue color**</font>.

```python
import numpy as np
import numpy.polynomial.legendre as lg
import matplotlib.pyplot as plt     
%pylab inline


def poly15(x):
    coef = [ 1.33164947 , -0.48120244, -1.93454292 , 0.79024579 , 2.33241528 , 0.27261564,\
              1.92751477 , 2.44851315 , 1.20674012,-3.04869176 , 0.21597876 , 1.08214072, \
              1.13487548 , 0.6082264 , -2.91413442]
    return lg.legval(x,coef)
    
xp = np.linspace(-1, 1, 100)
yp = poly15(xp)
plt.plot(xp,poly15(xp))

# Sampled
xsampled = np.linspace(-1, 1, 12)
ysampled = poly15(xsampled)
plt.scatter(xsampled,ysampled,c='b',marker='o',s=30)

plt.title('Test funcion: Polynomial of order 15')
```

We consider two models, or hypothesis, for learning our test function:
1. A polynomial of the second degree (the <font color="green">**green curve**</font>) 
2. A polynomial of order 12 (the <font color="red">**red curve**</font>). 

We will use ordinary least squares as our learning algorithm. The twelve order polynomial interpolates the data so its in-sample error is zero. However it is very wiggely while the in-sample error of the parabola approximation is far from zero but the approximation is very smooth and its out-of-sample error is much better that the  out-of-sample-error of the 12-order polynomial model.

```python
import warnings
warnings.simplefilter('ignore', np.RankWarning)

def outError(target,approx):
    M = 10000
    x  = numpy.random.uniform(low=-1.0, high=1.0, size=M)
    e  = np.mean( (target(x)-approx(x))**2 )
    return e

# Plot test function and sample points
plt.plot(xp,poly15(xp))
plt.scatter(xsampled,ysampled,c='b',marker='o',s=30)

# Hyphothesis H2 : Second order polynomial
p_coeff_2      = np.polyfit(xsampled,ysampled,2) 
poly_fit_2     = np.poly1d(p_coeff_2) 
plt.plot(xp,poly_fit_2(xp),'g',label='poly-2')

# Hyphothesis H12 : 12 order polynomial
p_coeff_12      = np.polyfit(xsampled,ysampled,12) 
poly_fit_12     = np.poly1d(p_coeff_12) 
plt.plot(xp,poly_fit_12(xp),'r',label='poly-12')
plt.legend()
plt.xlabel('x')
plt.title('Approximating 15th order polynomial')
plt.ylim([-2.,6.])

print 'Train set of 12 points\n',70*'-'
print 'Out-of-sample error for second order polynomial ',outError(poly15,poly_fit_2)
print 'Out-of-sample error for 12     order polynomial ',outError(poly15,poly_fit_12)
```

We will now increase the size of the data set from 12 to 120. Now we can see that the additional complexity of the 12-order polynomial model pays off and it achieves a much improved out-of-sample error as compared to the out-of-sample error of the second error polynomial model. 

```python
xsampled120     = np.linspace(-1, 1, 120)
ysampled120     = poly15(xsampled120)
p_coeff_2       = np.polyfit(xsampled120,ysampled120,2) 
poly_fit_2      = np.poly1d(p_coeff_2) 

p_coeff_12      = np.polyfit(xsampled120,ysampled120,12) 
poly_fit_12     = np.poly1d(p_coeff_12) 

print '\nTrain set of =120 points\n',70*'-'
print 'Out-of-sample error for second order polynomial ',outError(poly15,poly_fit_2)
print 'Out-of-sample error for 12     order polynomial ',outError(poly15,poly_fit_12)
```

## Lesson learned
>##<font, color="green">One should select the complexity of the model based on the size of the data set and not on the complexity of the test function that is approximated</font>

### <span id="OverfittingExampleLearningCurves">Detecting</span> overfitting using learning curves
In this demosntration the data set is composed of 60 points sampled from a polynomial of order 15. we compare the in-sample and out-of-sample errors by different polynomial models. a polynomial of order 12. As the order of the model increases the in-sample error always decreases **but** at some point the out-of-sample error starts to increase. This is the point where the problem of overfitting manifests itself.

```python
N           = []
Ein         = []
Eout        = []

target_func = poly15
deg         = 12

Ntest  = 10000
numpy.random.seed(seed=7)
perm   = np.random.permutation(Ntest)
xtest  = np.linspace(-1, 1, Ntest)[perm]
ytest  = target_func(xtest)[perm]

n_samples = 60
for deg in xrange(3,20,1):    
    xsample  =   xtest[:n_samples]
    ysample  =   ytest[:n_samples]
     
    p_coeff           = np.polyfit(xsample,ysample,deg) 
    poly_fit          = np.poly1d(p_coeff) 

    poly_fit_y_sample = poly_fit(xsample)
    poly_fit_y_test   = poly_fit(xtest)

    err_out = np.sum( (ytest-poly_fit_y_test)**2)/float(len(ytest))
    err_in  = np.sum( (ysample-poly_fit_y_sample)**2)/float(len(ysample))

    N.append( deg )
    Ein.append( err_in )
    Eout.append(err_out )
    
plt.plot(N,Ein, 'g',label='Ein')
plt.plot(N,Eout,'b',label='Eout')
plt.title('Approximation by polynom of order 15, 60 sample points')
plt.xlabel('Degree of approximating polynomial')
plt.ylabel('Square error')
plt.legend()
plt.ylim([0,16])
```

# <span id="TheCurseOfDimentionality">The</span> curse of dimensionality

Implementing machine learning methods becomes, sometimes exponentially harder, as the number of predictors, or attributes, increase. This [curse of dimentionality](http://en.wikipedia.org/wiki/Curse_of_dimensionality) is a problem that should be addressed whenever implementing a machine learning procedure. This is a serious problem that many times overlooked because our 3D-world intuition fails to carry us onto high dimensional spaces. In fact, most of the volume of high dimensional hyper-cubes is located at the boundaries as can be seen by the following examples: 
 * The volume of the unit sphere in $d$ dimensions is $\frac{2\pi^{d/2}}{\Gamma(n/2)}$ which goes to zero as $d \rightarrow \infty$
 * Thus, the ratio between the volume of a circimscribing circle and the unit cube and goes to zero with the dimension which means that most of the volume of the unit cube is concenrated at the "corners" of the unit cube
 * Suppose that we have point uniformly sampled and we ask what is the lenght $l$ of the side of a hyper-cube that contains 10% of the points. We get that $l=0.1^{1/d}$ so for $d=1$ we have $l=0.1$ and for $d=10$ we have $l=0.83$. That means that in order to cover 10% of the points we need the length of each side of the hyper-cube to cover 83% of the range in each direction. 
 * Local methods, e.g. KNN, are less local as the domension increase-EXPLAIN

There are also straighforward consequences of high dimensional spaces:
* The number of parameters of many learning methods grow with the dimension which leads to:
  * Increase in computation time for learning
  * Need to increase the size of the training set to avoid overfitting
* Number of training points have to grow exponentially with the dimension in order to keep a given sample density. For example, 10 points are needed to sample a unit interval so that adjacent points are at a distance of 0.1. If we want the same spacing between adjacent points for a $d$-dimensional cube then we will need $10^d$ points.
* This means that the training set will be sparse in high dimension and that the complexity of the learning model which usually grows with the dimension thus the chance for overfitting increases
* The notion of "nearest neighbor" in high dimension is very unintuitive


# <span id="ClassificationOfLearningMethods">Flavors</span> of learning methods

## Learning methods by available information
* [Supervised learning](http://en.wikipedia.org/wiki/Supervised_learning), also called inductive learning. In this type of learning the training data includes desired labels. This is the most studied part of machine learning. Note that as quantity of data grows it becomes more and more difficult to manually give every instance on the trainng set a proper label. Supervised learning can come in three main flavors:
  * **Classification** - The output of learner is discrete
  * **Regression** - The output of the learner is continuous
  * **Probability estimation** - The output of the learner is probability
* [Unsupervised learning](http://en.wikipedia.org/wiki/Unsupervised_learning) In this type of learning the training data does not include desired labels. 
* [Semi-supervised learning](http://en.wikipedia.org/wiki/Semi-supervised_learning) In this type of learning the some of the training samples include the desired labels. 
* [Active learning](http://en.wikipedia.org/wiki/Active_learning_(machine_learning)) - A special case of semi-supervised learning in which the learning machine can requests the user to provide labels for desired samples.
* [Reinforcement learning](http://en.wikipedia.org/wiki/Reinforcement_learning). Kind of a play where the player takes an action and is rewarded for success or punished for failure

## Modeling approaches for classification  
* **Generative models** Compare the input to the *best approximation of the target function* and sets the paramaters to maximize the likelihood of observed training data
* **Discriminative models** Compare the input to the *best approximation of the boundary between classes* and sets the paramaters to maximize the likelihood of observed training data
* **Nearest Neighbor models** compare the input to (nearest) training data and vote. Also called **instance** based learning. Examples: k-nn, kernel methods

https://depts.washington.edu/uwcl/twiki/pub/Main/TopicsDiscussed/stat-classification.pdf

## Learning methods by distribution of work between learning and prediction step 
* [Eager learning](http://en.wikipedia.org/wiki/Eager_learning) - The system tries to construct a general, input independent target function during training of the system. This is an example of the **offline learning** approach in which the system do not change its approximation to the target function once the initial training phase is completed. The training phase is the part where the best instance of the hypothesis space is selected based on the available data.
* [Lazy learning](http://en.wikipedia.org/wiki/Lazy_learning) - generalization beyond the training data is delayed until a query is made to the system, also instance based larning
* [online learning](http://en.wikipedia.org/wiki/Online_machine_learning) - methods that assume that the data comes in a **sequential** (or **streaming**) form, This form of learning is useful for the "big data" regime where the data can not be stored in memory of single disk and when the predictin model have to change over time. The case where the entire training set can be accessed is called **batch learning**. There is also a **mini-batch** scenario where parts of the training set are available. 

## Search procedure
How the optimization algorithm goes about searching for good hypothesis:
* **One step computation** - Solving for theparameters in one step. Examples: Linear regression, Naive Bayes,.. 
* **Local search** - start with an inuitial hypothesis, making small improvments until reaching a local minimum. Example: (Stochastic) Gradient descent for Neural Network 
* **Constructive search** - Start with an empty hypothesis, gradually add more structure to it untill reaching a local minimum. Example: Decision trees, Boosting

# <span id="TheoryLearning">Theories</span> of learning
The term "learning" is actually referred to **generalization**. We would like our selected hypothesis $g^{(D)}$ to perform well on data it has not seen. We would like the **in-sample** error $E_{\tt in}(g^{(D)}) $ to be close to the **out-of-sample** error $E_{\tt out}(g^{(D)})$. In other words, we would like $||R(g^{(D)}) - R_{\tt{emp}}(g^{(D)}) ||$ ) to be small. There are some interesting topics covered in the theory of learning such as
* Is learning feasible?
* How many sample points do we need to get good generalizations?
* What learning machine should we use - what are the tradeoffs?

There are more than one learning theory covering the questions above from different viewpoints. Those theories arer beatiful in their own right and provide many useful hints for selecting, using and evaluating a learning method. My very informal summary of the method is that  
1. Learning from data is, in general, impossible if we are not given any additional information. This is the consequence of the no-free-lunch theorems. Fortunatly real learning problems are never stated in the general terms of this theorem and the implicit end explisit restrictions make learning possible
2. Second result is that the motre powerful a learner is, the more sample points are needed fro good geenralizations. Providing too few sample points fro powerful learners causes **overfitting**. This is expressed, from different angles by PAC learning, Bias-Variance decomposition of th egeneralization error and theory of VC dimension.

## <span id="NoFreeLunchTheories">No</span> free lunch theorems

[No free lunch theorems](http://en.wikipedia.org/wiki/No_free_lunch_theorem). We consider the statement for a simple case of binary classification model (that is, concept learning) based on binary predictors. Let  $\mathcal{F}$ be the set of all boolean functions. Then:
>  $\frac{1}{|\mathcal{F}|} $$\sum_{f \in \mathcal{F}} E_{\tt{out}}(g,f) = \frac{1}{2}$ 

where $E_{\tt{out}}(g,f)$ is the out-of-sample error (fraction of instances of the test set that are wrongly classified) for the best hypothesis $g$ consructed by the learner $L$ over the training set $D$. This is true for any training set and learner. 

The proof is very simple: Suppose that we have a learner that achieves a generalization error of $\frac{1}{2}-\epsilon$ for a given target function $f$. then there is a target function $f^*$ that is equal to $f(x)$ when $x$ is in the training data and is equal to $\neg f(x)$ when $x$ is in the test set. Then the generalization error for $f^*$ would be $\frac{1}{2}+ \epsilon$. Summing over all pairs of functions, we get the result.

This suggests that no learner can in principle be better than a random coin-flip learner. Thus no learning model or inductive bias is better than others. Of course, in reality a target function is not randomply selected out of the space of target functions which is the main assumption. In practice we impose other (unknown but not uniform) probability distribution for specific problem domaions, exploiting regularities, smoothness and other domain characteristics. However the theory do explain, at least qualitativly, why the premise of the single jack-of-all-trades algorithm in machine learning was never fulfiled. In many cases simple algorithms outperformed more complex alternatives in specific problem domains. The NFL theorems also explain the need for a large tool box of different algorithms in order to be a good practitionner in machinbe learning.


## <span id="BiasVariance">Bias</span>-Variance tradeoff
The notion of **Bias** and **Variance** are welll established in the machine learning community and are used to explain, evaluate and compare the performance of various algorithms. Informaly speaking, the out-of-sample error of a learner is decomposed into three parts: The **Bias** represent how well the best predictor from our hypothesis set fits the target function, on average for many randomly selected trainuing sets of the same size. The **Variance** represents how **stable** is the optimal predictor when we vary the data set. Finally the **irreducible error**, sometimes loosly called the **noise** represent the parts of the target function that can not be approximated by our hypothesis set. 

This notion was found to be useful in developing and evaluating machine learning algorithms: While the Noise term is outside of our control, we have control over the Bias and Variance. Learners with low bias and high variance are the flexible models that can approximate well complex target functions but are unstable and may vary widely from data set to data set. The models with high bias and low variance are the opposite: They have low approximation power but are very stable and do not deviate much from one dataset to another.


### Bias-Variance decopmposition for square loss

We consider an hypothesis set $\mathcal{H}$ from which the learning algorithm selects the best hypothesis $g^{(D)}=g^{(D)}(x)$. The Bias-Variance decompositions is the following: 
> $\mathbb{E}_D[E_{\tt out}(g^{(D)})]= \tt{Bias}^2 + \tt{Variance} + \tt{Noise}$
 
where $\mathbb{E}_D[E_{\tt out}(g^{(D)})]$ is the average out-of-sample error over all data sets of size $N$, that is
> $E_{\tt out}(g^{(D)}) = E_x[(g^{(D)}-y)^2]$ 

where $ y = f(x) + \epsilon$ and $\epsilon$ is a "noise" term with zero mean ($E_D[\epsilon]=0$) and variance $\tt{Var}(\epsilon)=E_D[\epsilon^2]$ independent of $x$.
Here $E_x$ denots averaging over the test set (this is just the definition of the risk) and thus
>$E_D[E_{\tt out}(g^{(D)})] = E_x[E_D[(g^{(D)}-y)^2]]$. 

To evaluate the expectation $E_D [ (g^{(D)}-y)^2 ]$ we will use the 'average' hypothesis $\bar{g} = E_D[g^{(D)}]$ where, for each x, we average over **many** data sets $D_1,D_2,...D_K$ so  $\bar{g} = \bar{g}(x) = \frac{1}{K} \sum_{i=1}^K g^{(D_K)}$. Using $\bar{g}$ we get the decomposition:

>$E_D [ (y-g^{(D)})^2 ] = E_D [ ( y-f + f - g^{(D)} )^2 ] \\
\quad = E_D [( y-f)^2] + E_D [(f-g^{(D)} )^2 ] + 2 E_D [(y-f)(f-g^{(D)} ) ] \\
\quad = \tt{Var}(\epsilon) + E_D [(f-g^{(D)} )^2 ] \\
\quad = \tt{Var}(\epsilon) + E_D [(f-\bar{g}+\bar{g}-g^{(D)} )^2 ] \\
\quad = \tt{Var}(\epsilon) + E_D [(f-\bar{g})^2] + E_D [(\bar{g}-g^{(D)} )^2] + 2E_D [(f-\bar{g}))(\bar{g}-g^{(D)} )] \\
\quad = \tt{Var}(\epsilon) + E_D [ (g^{(D)}-\bar{g})^2 ] + (\bar{g}-f )^2 \\  
\quad \equiv \tt{Var}(\epsilon)  + {\tt var}(x) + {\tt bias}^2(x)$ 

where we note that the cross terms $2 E_D [(y-f)(f-g^{(D)} ) ]$ and $2E_D [(f-\bar{g}))(\bar{g}-g^{(D)} )] $ are zero. This is because  $E_D[y]=E_D[f]=f$, $\bar{g}$ is independent of $\epsilon$ and thus $E_D[\epsilon \bar{g}] = E_D[\epsilon] E_D[\bar{g}]=0$,  $E_D[g^{(D)}]=\bar{g}$ and $(f-\bar{g})$ does not depent on $D$. Integrating over the domain $\mathcal{X}$, we get the Bias-Variance decomposition:

> $E_D[E_{\tt out}(g^{(D)})] = \tt{Var}(\epsilon) + E_x[{\tt var}(x)] + E_x[{\tt bias}^2(x)] = \tt{Noise}  + \tt{Variance} + {\tt Bias}^2$

### Bias-Variance decopmposition for other loss functions
The Bias-Variance decomposition arose from regression probelms where the square loss dominates and it proved to be very successfull and intuitive. Extending Bias-Variance to other loss functions and in particular to the 0-1 loss function (frequency of prediction errors) of classification problems turned out to be less obvious. In fact Gareth James in his paper entitled *"Variance and bias for general loss functions"* ([download](http://www-bcf.usc.edu/~gareth/research/bv.pdf)) argued that there are two characteristics of a p;lausible definitions of a Bias-Variance decomposition that coincide for the square loss functions but can not be simultanously met for the 0-1 loss. These are: (a) providing and additive decomposition of the predictive error and (b) reproduction of the standard characteristics of BIas and Variancem e.g. that a variance cannot be negative. We will not discuss this issue further in this section. The interested reader is encouraged to look at some papers listed in the <a href="#LearningTheoryResources" data-ajax="false">Resources</a> <span id=""> section.

```python
import numpy as np
from IPython.html.widgets import interact, interactive, fixed
from IPython.html import widgets

import random
import numpy.polynomial.legendre as lg
import matplotlib.pyplot as plt     ## matplotlib - plots
#import seaborn as sns
%pylab inline

def polyWithNoise(x,err_factor):
    coef5 = [ 2.33164947 , -0.48120244, -1.93454292 , 0.79024579 , 2.33241528]
    p     = lg.legval(x,coef5)
    p += err_factor*np.random.randn(len(x))
    return p

def poly_fit_deg(xp,xsampled,ysampled,deg):
    p_coeff      = np.polyfit(xsampled,ysampled,deg) 
    poly_fit     = np.poly1d(p_coeff) 
    plt.plot(xp,poly_fit(xp),'r',label='hypothesis')


def plotOverfit(n_sample,deg,show_data_points):
    # PLot target function
    err_factor = 0.0
    xp = np.linspace(-1, 1, 100)
    yp = polyWithNoise(xp,err_factor)
    plt.plot(xp,yp,'g',label='target function')

    # Sample n_samples points
    err_factor = 0.3
    xsampled = np.linspace(-1, 1, n_sample)
    ysampled = polyWithNoise(xsampled,err_factor)
    
    if show_data_points:
        plt.scatter(xsampled,ysampled,c='b',marker='o',s=30)

    # Fit polynomial by OLS
    poly_fit_deg(xp,xsampled,ysampled,deg)

    plt.title('Approximating polynomial of order 4')
    plt.ylim([0.,5.])
    plt.legend()


def show_args(**kwargs):
    plotOverfit(kwargs['n_samples'],kwargs['deg_hypothesis'],kwargs['show_data_points'])

i = interact(show_args,
         show_data_points   = True,
         c = widgets.ToggleButtonWidget(description="Run"),
         a=widgets.FloatSliderWidget(min=2, max=32, step=1, value=2, description="deg_hypothesis"),
         b=widgets.FloatSliderWidget(min=12, max=240, step=5, value=12, description="n_samples")      
         )
```

## <span id="PACLearning">PAC</span> learning

There are many versions of the theory of Probably Approximatly Correct ([PAC](http://en.wikipedia.org/wiki/Probably_approximately_correct_learning)) learning. Jeremy Kun wrote a very [nice post](http://jeremykun.com/2014/09/19/occams-razor-and-pac-learning/) describing PAC learning. PAC learning is an exciting theory because: (a) It provides assurances for inductive learning and (b) It gives us some guidlines on the number of examples needed to learn. 

In this section I will summarize, somewhat informaly, the main results. For simplicity, I will use the notations of this notebook that are different from the traditional notations of PAC learning. 
We will present the simplest possible setting, restricting our attention to concept learning, namely a binary classification problems. We consider a binary target function $f(x)$. In the PAC setting we give two constants $ 0 < \epsilon, \delta < 0.5$ and we search for the number $N$ of samples in the training set so 
> $P( E_{\tt{out}}(g) > E_{\tt{in}}(g)  + \epsilon )$

is small.. that is the **probability** that the out-of-sample error will be **approximatly** equal to the in-sample can be made arbitrarily small, provided we are willing to provide many examples.

* **Finite dimensional hypothesis space with $E_{\tt{in}}(g)=0$**

  **Claim**
  
   >  <font color="green">$\delta \equiv  P( E_{\tt{out}}(g) \geq \epsilon ) \leq |\mathcal{H}|e^{-N\epsilon} \Longrightarrow N \leq  \frac{1}{\epsilon}( \log(|\mathcal{H}|) + \log(\frac{1}{\delta}))$</font>
    
  **Proof**
    
    We define a "bad" hypothesis $h$ to be an hypothesis for which $E_{\tt{out}}(h) > \epsilon$. This means that the proportion of points badly labeled is greater that $\epsilon$. The probability that a bad hypothesis is consistent is
    >$P(E_{\tt{in}}(h)=0) = P(\cap_{i=1}^N\{h(\mathbf{x_i})=y_i\}) = \Pi_{i=1}^N P(\{h(\mathbf{x_i})=y_i\})\leq (1-\epsilon)^N$
    
  $P(E_\tt{out}(g) > \epsilon)$ is the probability that we learned a bad hypothesis.  This probability is not greater than the probability that the set of consistent hypothesis contains a bad hypothesis. By the [Union=Bound](http://en.wikipedia.org/wiki/Boole%27s_inequality) inequality to get
    
  > $ P(E_\tt{out}(g) \geq \epsilon) < S (1-\epsilon)^R$ where $S$ is the number of consistent hypothesis
    
   now we bound $S$ by $|\mathcal{H}|$, the cardinality of the hypothesis set $\mathcal{H}$ and use the inequality $1-\epsilon \leq e^{-\epsilon}$ to get
   
   > $\delta \equiv P(E_\tt{out}(g) > \epsilon) \leq |\mathcal{H}|e^{-N\epsilon}$
.


* **Finite dimensional hypothesis space with $E_{\tt{in}}(g)>0$**
  
  This case is also called **agnostic** learning since we wish to get results even when the target function is not within the hypothesis set. Clearly we can not hope for the stron assurances of the first case:

    **Claim**
  
  >  <font color="green">$\delta \equiv P( E_{\tt{out}}(g) > E_{\tt{in}}(g)  + \epsilon ) \leq |\mathcal{H}|e^{-2N\epsilon^2} \Longrightarrow  N \leq  (2/\epsilon^2)( \log(|\mathcal{H}|) + \log(\frac{1}{\delta}))$</font>
    
  **Proof**
   
    In this case we use [Hoffeding's Inequality](http://en.wikipedia.org/wiki/Hoeffding%27s_inequality) that states that for any hypothesis $h$, we have that $P( E_{\tt{out}}(h) > E_{\tt{in}}(h)  + \epsilon ) \leq e^{-2N\epsilon^2} $. The rest follows from Union-Bound.


* **Infinite dimensional hypothesis space with $E_{\tt{in}}(g)>0$**

    The key idea here is the theory of [VC-dimension](http://en.wikipedia.org/wiki/VC_dimension) which allows us to replace $\log(|\mathcal{H}|)$ by  $\tt{VC}(\mathcal{H})$, the VC-dimension of th e hypothesis space $\mathcal{H}$. THe VC dimension alows us to reason about the complexity, or prediction power of the hypothesis space. The larger is the VC-dimension, the more is the predictive power of $\mathcal{H}$ wbut, also, the more examples we will need to allow generalization. The VC-dimension is the largest index $k$ that allow us winning the following game: We can choose to select any points $\mathbf{x_1},\mathbf{x_2},\ldots,\mathbf{x_k}$ and we win the game if we can find a consistent learner (one that attains zero in-sample error) $g \in \mathcal{H}$ for any labels but the teacher can assign **any** labels $y_1,y_2,\ldots,y_k$. If we can find a consistent learner for any number of sample points, the the VC-dimension is infinite. As an example we note that he VC-dimension of a linear hypothesis fro $d$ predictors in $d+1$ and the VC-dimension for decision tree is infinite since this is a very flexible learnin model. Note that in some cases he VC-dimension is assotiated with the number of parameters of a model but this is not true in general. THere are models with many parameters with a small VC-diomension and a famous example of a model with just one parameter having infinite VC-dimension.
    
  **Claim**
  
  >  <font color="green">$ N \leq  (1/\epsilon)( 8\tt{VC}(\mathcal{H})\log_2(13/\epsilon) + 4 \log_2(2/\delta))$</font>

  **Consequence**
  
  Learning (generalization) is only possible if the VC-dimension is finite.


It is indeed exciting the the theory of PAC learning is able to provide bound on the number of samples needed in a very general setting. However, the bounds obtained by the theory are far too conservative to be of practical. Nevetheless the information provides us good understanding of the tradeoffs involved, mainly between the **complexity** of the learner and the number of data points needed to achieve good generalizations.

### <span id="PACLearningExample">Example</span> PAC learning
In this demosntration we compare the in-sample and out-of-sample errors to a 15th order polynomial that is approximated by a polynomial of order 12.

```python
import numpy as np
import numpy.polynomial.legendre as lg
import matplotlib.pyplot as plt     ## matplotlib - plots
import warnings
warnings.simplefilter('ignore', np.RankWarning)


#import seaborn as sns
%pylab inline


def poly15(x):
    coef = [ 1.33164947 , -0.48120244, -1.93454292 , 0.79024579 , 2.33241528 , 0.27261564,\
              1.92751477 , 2.44851315 , 1.20674012,-3.04869176 , 0.21597876 , 1.08214072, \
              1.13487548 , 0.6082264 , -2.91413442]
    return lg.legval(x,coef)
    

N           = []
Ein         = []
Eout        = []

target_func = poly15
deg         = 12

Ntest  = 10000
numpy.random.seed(seed=7)
perm   = np.random.permutation(Ntest)
xtest  = np.linspace(-1, 1, Ntest)[perm]
ytest  = target_func(xtest)[perm]


for n_samples in xrange(3,1320,1):    
    xsample  =   xtest[:n_samples]
    ysample  =   ytest[:n_samples]
     
    p_coeff           = np.polyfit(xsample,ysample,deg) 
    poly_fit          = np.poly1d(p_coeff) 

    poly_fit_y_sample = poly_fit(xsample)
    poly_fit_y_test   = poly_fit(xtest)

    err_out = np.sum( (ytest-poly_fit_y_test)**2)/float(len(ytest))
    err_in  = np.sum( (ysample-poly_fit_y_sample)**2)/float(len(ysample))

    N.append( n_samples )
    Ein.append( err_in )
    Eout.append(err_out )
    
plt.plot(N,Ein, 'g',label='Ein')
plt.plot(N,Eout,'b',label='Eout')
plt.title('Learning curve: Approximation by polynm of order 12')
plt.xlabel('Number of training points')
plt.ylabel('Square error')
plt.legend()
plt.ylim([0,6])
```

### <span id="PopularLossFunctionRegression">Popular</span> loss functions for regression

The lost function is a function of one variable $t=w-y$. We consider the functions

* Square loss  
   >$L(w,y) = (w-y)^2$. 
   
   In this case the optimal estimator is given by the conditional expectation $f_o(x) = E[Y \mid X=x]$, also called a **regressor** in the statistics literature.
* Absolute loss 
> $L(w,y) = |w-y|$
  
  In this case the optimal estimator is given by the conditional median $f_o(x) = \tt{median}(Y \mid X=x)$.
* $\epsilon$-insensitive loss 
  > $L(w,y) = \max \{ |w-y|-\epsilon,0 \}$

All the above loss functions are convex.


### <span id="PopularLossFunctionClassification">Popular</span> loss functions for classification

We assume the labels $y$ to be from the set $\{1,-1\}$. The loss function is a function of one variable $t=wy$ and we consider the following functions
* Square loss which is the convex loss function 
  > $L(w,y) = (w-y)^2 = (1-wy)^2$

* [Hinge loss](http://en.wikipedia.org/wiki/Hinge_loss) is the loss function related to the maximal margin principle behing Support Vector Machines classifiers.
  >$L(w,y) = \max \{ 1-wy,0 \} =|1-wy|_{+}$

* [Logistic loss](http://www.hongliangjie.com/wp-content/uploads/2011/10/logistic.pdf), or cross entropy loss, is the convex loss functiona that leads to logistic regression
  > $L(w,y) =  \log(1+e^{-wy})$

* [Exponential loss](http://en.wikipedia.org/wiki/AdaBoost) which is the loss of the boosting learning algorithm 
  > $L(w,y) = e^{-wy}$ 

* The 0-1 loss 
  > $L(w,y) = I(w \neq y) = I(wy<0) $
  
  that counts the number of missclassified points. Here $I(\cdot)$ denotes the indicator function that has value 1 if the condition is true (predicted value equal to data value) and zero otherwise. 
  
  The optimal solution for the 0-1 loss function is $f_o ({\mathbf x})=g_{k({\bf x})}$ where $g_{k({\bf x})}$ such that $ P(Y=g_{k({\bf x})} \mid X={\mathbf x})= \max_{g_i \in \{1,-1\}} P(Y=g_i \mid X={\mathbf x})$. This solution is known as the **Bayes classifier**

  This loss function is not smooth and not convex and leads to intactable optimization problems. Thus many times we use **surrogate loss function** that are convex approximation to the 0-1 loss. Such are the hinge loss, logistic loss and the exponential loss. {SEE ALSO](http://www.cs.utah.edu/~piyush/teaching/13-9-print.pdf)


# <span id="CombatingOverFitting">Combating</span> overfiting

[Overfitting](http://en.wikipedia.org/wiki/Overfitting) is probably the most serious problem whne applying a machine learning technique to a real problem. This term is a **comparative** terms and refers to the situation where the in-sample of one alternative reduces while the out-of-sample error increases. Overfitting manifests itself when the number of samples does not match the complexity of the model. In terms of Bias-Variance analysis, overfitting is the case where the bias is decreased because of the flexibility of the model but the variance is very high. In terms of PAC learning, overfitting happen when the increase complexity of the model is not respected by suppying more data points. Some view overfitting as a phenomenom in which the model adapts itself to the noise in the data and it halucinates the target function instead of genealizing. 

In his course, Yaser Abu-Mostafa gives a rule of thumb to the minimal number of samples that we should have in order to provide generalization and avoid overfitting: <font color="green">The number of samples should be $\geq 10 \cdot \tt{VC}(\mathcal{H})$</font>. In practice, a great deal of though are invested in any machine learning algorithm to handle and minimize the risk of overfitting.



### <span id="regularization">Regularization</span>

[Regularization](http://en.wikipedia.org) is a common apparoach to tackle overfitting. We can arrive to reggularization from many different angles: The classical regularization theory aims at correcting the inherent ill-posenet o fth elearning problem. In the present context, it is more natural to me to think o fthe problem from the PAC learning viewpoijint. We want to decrease the variance of our model while not increasing the variance too much. WeOne of the cures for overfitting is to reduce the VC dimension of $\mathcal{H}$. One way of accomplishing this is to restrics the size of the coefficient vector, One way is by [feature selection](http://en.wikipedia.org/wiki/Feature_selection),that is reduce the number of non-zero coefficients. This leads to intractable optimization problem and thus we replace that by requesting that $||\beta|| < C$ for some constant $C$. Small values of $C$ correspond to large regularization while large values of $C$ result is small regularization. 

There are two equivalent formulations for regularization in the linear case. The first is based directly on the desire to restrics the value of $\beta$

>$\min_{||\beta||^2 < C} ( X\beta - {\mathbf y})^T( X\beta - {\mathbf y})$

We note that in the formulation above we can replace $||\beta||^2 = \sum_{i=1}^{p} \beta_i^2$ by $||\beta||^2 = {\mathbf \beta^T \beta} = \sum_{i=1}^{p} \gamma_i \beta_i^2 = \beta^T \Lambda \beta$ where $\Lambda$ is a diagonal matrix or the more general [Tikhonov regularization](http://en.wikipedia.org/wiki/Tikhonov_regularization) $\beta^T \Gamma^T \Gamma \beta = ||\Gamma \beta||^2$

The second for for the regularization problem, obtained from the first via Lagrange multipiers, is expressed as **uncponditional** minimization problem taht, for Tikhonov regularization, is
> $\min ( X\beta - {\mathbf y})^T( X\beta - {\mathbf y}) +||\Gamma \beta||^2$


where, in th simplest formulation $\Gamma = \lambda I$ and  $\lambda$ is inversly proportional to $C$. The above problem has a one step solution $\beta = (X^TX + \Gamma^T \Gamma)^{-1} X^Ty$

This approach to regularization is also called **weigh decay** (for neural networks) or **shrinkage** (in statistical literature).

We note that we can regularize using other norms. In particular the $L_1$ norm, called **lasso** in the statistical literature is attractive as it produces sparse models and thus do also feature selection in the process of regularization. However, this penalty is not differentiable which complicates optimization algorithms.

## <span id="regularizationExamplePolynomial">Regularization</span> example - polynomial
Fit a polynomial of degree 15 (blue curve) over a sample of 12 points using ordinary least squares (red curve) and ridge regression learning algorithm with two choices of the smoothing parameter (solid and dashes green curves).

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.grid_search  import GridSearchCV


def createTrainingSet(deg,xsampled):
    N = len(xsampled)
    X = np.zeros((N,deg+1))
    for n in xrange(N):
        X[n] = np.array([ np.power(xsampled[n],i) for i in xrange(deg+1) ] )
    return X

def outError1(target,est,deg):
    M = 10000
    e = 0.0
    for m in xrange(M):
        x     = random.uniform(-1.,1.)
        xp    = [np.power(x,i) for i in xrange(deg+1)]
        ypred = est.predict(xp)
        e    += (target(x)-ypred)**2
    e /= float(M)
    return e

#==========================================================================
# Generate train set
#==========================================================================
xtrain = np.linspace(-1, 1, 12)
ytrain = poly15(xsampled)
deg    = 12
X      = createTrainingSet(deg,xtrain)

#==========================================================================
# Fit
#==========================================================================
#--------------------------------------------
# Fit with linear regressor
#--------------------------------------------
est_lin   = LinearRegression(fit_intercept=False)
est_lin.fit(X,ytrain)

#--------------------------------------------
# Fit with ridge regressor
#--------------------------------------------
est_ridge = Ridge(fit_intercept=False)
lamda = 500.
est_ridge.fit(X,ytrain,lamda)


#--------------------------------------------
# Fit with ridge regressor and auto determination
# of alpha
#--------------------------------------------
print '\nDetermine alpha by cross-validation\n',50*'-'
alphas_to_try = np.arange(0.1,10000,20)
model = Ridge(fit_intercept=False)
grid  = GridSearchCV(estimator=model,param_grid=dict(alpha=alphas_to_try))
grid.fit(X,ytrain)
print 'Best alpha found is ',grid.best_estimator_.alpha

est_ridge_auto = Ridge(fit_intercept=False)
est_ridge_auto.fit(X,ytrain,grid.best_estimator_.alpha)

#==========================================================================
# Print out-of-sample errors
#==========================================================================
print '\nTrain set of 12 points. Fitting twelve order polynomial\n',70*'-'
print 'Out-of-sample error polyfit by linear regression                      ',\
       outError1(poly15,est_lin,deg)
print 'Out-of-sample error polyfit by ridge  regression (alpha=500)          ',\
       outError1(poly15,est_ridge,deg)
print 'Out-of-sample error polyfit by ridge-reg alpha=20 (cross-validation)  ',\
       outError1(poly15,est_ridge_auto,deg)


#==========================================================================
# Plot all
#==========================================================================
xp = np.linspace(-1, 1, 100)
yp = poly15(xp)
plt.plot(xp,poly15(xp))                                 # Plot target function

plt.scatter(xsampled,ytrain,c='b',marker='o',s=30)    # Plot training set


yfit_lin = []
for xx in xp:
    yfit_lin.append(est_lin.predict([np.power(xx,i) for i in xrange(deg+1)]))
plt.plot(xp,yfit_lin,'r')

yfit_ridge = []
for xx in xp:
    yfit_ridge.append(est_ridge.predict([np.power(xx,i) for i in xrange(deg+1)]))
plt.plot(xp,yfit_ridge,'g')

yfit_ridge_auto = []
for xx in xp:
    yfit_ridge_auto.append(est_ridge_auto.predict([np.power(xx,i) for i in xrange(deg+1)]))
plt.plot(xp,yfit_ridge_auto,'g--')


plt.ylim([-2.,6.])

```
